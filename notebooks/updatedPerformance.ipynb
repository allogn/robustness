```python
import matplotlib.pyplot as plt
import pandas as pd
import pylab
%matplotlib inline

import sys, os
sys.path.append(os.path.dirname(os.path.abspath('')))
sys.path.append(os.path.join(os.path.abspath(''),'..','scripts'))
sys.path.append(os.path.join(os.path.abspath(''),'..','src'))

from Experiment import *
from Analyzer import *
from Artist import *
```

```python
analyst = Analyzer("emt_performance")
results = analyst.df
```

```python
results = results.sort_values("n")
# fig, ax = Artist().create_figure()
sns.set(style="whitegrid", font_scale=1.8, rc={'axes.grid': True})
fig = plt.figure(figsize=(5.5, 2.9), dpi=150)
ax = fig.add_subplot(1, 1, 1)

lmap = {0: "DynIM", 1:"DFS", 2:"TD-SCC", 3:"BU-SCC", 4:"DynSCC", 5:"S-Dagger", 6:"BIT-Dagger"}
markers_and_lines = [('--','o'),('-','v'),('-.','^'),(':','D'),('-','s'),('--','h')]
i = 0
lines = []
for mode in [1,2,3,4,5,6]:
    lines += ax.plot(results[results['mode'] == mode]["n"], 
             results[results['mode'] == mode]["runtime"]/(10**9),label=lmap[mode], 
             linestyle=markers_and_lines[i][0], marker=markers_and_lines[i][1])
    i += 1
ax.set_yscale("log")
ax.set_xlabel("n")
ax.set_ylabel("Runtime [s]")
# ax.legend(bbox_to_anchor=(1.05, 1), loc='upper left', borderaxespad=0.)
# ax.legend(bbox_to_anchor=(1., 1.02, 0., .102), loc=3,
#            ncol=2, mode="expand", borderaxespad=0.)
# ax.legend(bbox_to_anchor=(-.35, -0.4, 1.6, .102), loc='upper left',
#            ncol=3, mode="expand", borderaxespad=0.,frameon=True)

# ax.legend(bbox_to_anchor=(-.35, -0.4, 3, .102), loc='upper left',
#            ncol=6, mode="expand", borderaxespad=0.,frameon=True)
```
